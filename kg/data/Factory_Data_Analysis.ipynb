## 9. Plot different features against one another (scatter), against frequency (histogram)

### Histogram

Frequency of occurrence of variables in an interval.

```
df.factory.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Records by FactoryId")
plt.ylabel('No. Records')
plt.xlabel('Factory');
```

### Heat Maps

```
#ndf = df.drop(['date','factory','location','machine_type', 'defect_root_cause','raw_material_quality','maintenance_history'], axis=1)
ndf = df.select_dtypes(exclude=['object'])
ndf.head(5)
```

```
plt.figure(figsize=(25,30))
c= ndf.corr()
sns.heatmap(c,cmap="BrBG",annot=True)
c
```

### Scatterplot

scatter plots to find the correlation between two variables

```
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(df['defect_rate_pcnt'], df['profit_margin_pcnt'])
ax.set_xlabel('defect_rate_pcnt')
ax.set_ylabel('profit_margin_pcnt')
plt.show()
```
